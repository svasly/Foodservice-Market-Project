# Описание данных

Файл `/datasets/rest_info.csv` содержит информацию о заведениях общественного питания:
- `name` — название заведения;
- `address` — адрес заведения;
- `district` — административный район, в котором находится заведение, например Центральный административный округ;
- `category` — категория заведения, например «кафе», «пиццерия» или «кофейня»;
- `hours` — информация о днях и часах работы;
- `rating` — рейтинг заведения по оценкам пользователей в Яндекс Картах (высшая оценка — 5.0);
- `chain` — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
- `0` — заведение не является сетевым;
- `1` — заведение является сетевым.
- `seats` — количество посадочных мест.
Файл `/datasets/rest_price.csv` содержит информацию о среднем чеке в заведениях общественного питания:
- `price` — категория цен в заведении, например «средние», «ниже среднего», «выше среднего» и так далее;
- `avg_bill` — строка, которая хранит среднюю стоимость заказа в виде диапазона, например:
- «Средний счёт: 1000–1500 ₽»;
- «Цена чашки капучино: 130–220 ₽»;
- «Цена бокала пива: 400–600 ₽». и так далее;
- `middle_avg_bill` — число с оценкой среднего чека, которое указано только для значений из столбца `avg_bill`, начинающихся с подстроки «Средний счёт»:
- Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
- Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
- Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
- `middle_coffee_cup` — число с оценкой одной чашки капучино, которое указано только для значений из столбца `avg_bill`, начинающихся с подстроки «Цена одной чашки капучино»:
- Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
- Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.

```python
def show_missing_stats(tmp0):
    """
    Функция для отображения статистики пропущенных значений в DataFrame.
    """
    missing_stats = pd.DataFrame({
        'Кол-во пропусков': tmp0.isnull().sum(),
        'Доля пропусков': tmp0.isnull().mean()
    })
    missing_stats = missing_stats[missing_stats['Кол-во пропусков'] > 0]
    
    if missing_stats.empty:
        return "Пропусков в данных нет"
    
    # Форматируем при выводе через Styler
    return (missing_stats.style.format({'Доля пропусков': '{:.4f}'}).background_gradient(cmap='coolwarm'))
show_missing_stats(df)
```

Для дальнейшей работы создадим столбец `is_24_7` с обозначением того, что заведение работает ежедневно и круглосуточно, то есть 24/7:
  - логическое значение `True` — если заведение работает ежедневно и круглосуточно;
  - логическое значение `False` — в противоположном случае.

```python
display("Создание столбца is_24_7")

def check_24_7(hours_str):
    """
    Проверяет, работает ли заведение круглосуточно.
    Возвращает True, если заведение работает 24/7.
    """
    if pd.isna(hours_str):
        return False
    
    hours_lower = str(hours_str).lower()
    
    # Список паттернов, указывающих на круглосуточную работу
    patterns = [
        '24/7',
        'ежедневно, круглосуточно'
        ]
    
    for pattern in patterns:
        if pattern in hours_lower:
            return True
    
    return False

df['is_24_7'] = df['hours'].apply(check_24_7)
```

```python
# Проверим правильность категоризации
tmp0, col1 = df.copy(), 'is_24_7'
a1 = len(tmp0[tmp0.hours != 'ежедневно, круглосуточно'])
a2 = len(tmp0[tmp0.hours == 'ежедневно, круглосуточно'])

print(" Количество не работающих ежедневно и круглосуточно", a1,
      '\n', "Количество работающих ежедневно и круглосуточно", a2,'\n')
print("Количество категорий полученное в проекте:")
try:
    print(tmp0[col1].value_counts())
except:
    print("Не создан столбец с бинарным признаком ежедневной и круглосуточной работы заведения.")
```

### Задача 6

Изучим, с какими данными показывают самую сильную корреляцию рейтинги заведений. Построим и визуализируем матрицу корреляции рейтинга заведения с разными данными: его категорией, положением (административный район Москвы), статусом сетевого заведения, количеством мест, ценовой категорией и признаком, является ли заведения круглосуточным. Выберем самую сильную связь и проверим её.

**6.1. Подготовка данных для корреляционного анализа**

```python
# Создаем копию данных только для корреляции
df_phik = df[['category', 'district', 'rating', 'chain', 'seats', 'price', 'is_24_7']].copy()

# Возвращаем заглушку 'не указано' обратно в NaN
# Это необходимо, чтобы при расчете корреляции эта группа не учитывалась как отдельная категория
df_phik['price'] = df_phik['price'].cat.remove_categories('не указано')

# Проверяем результат подготовки
print("Размер данных для корреляции:", df_phik.shape)
print("Количество пропусков в price после замены:", df_phik['price'].isna().sum())
```

**Промежуточный вывод 6.1:**

Данные подготовлены для корректного расчета φ-коэффициента. 
Заглушка "не указано" заменена на NaN, что позволит библиотеке phik не учитывать эти записи 
как отдельную категорию при расчете корреляции.

**6.2. Расчет матрицы корреляции φ-коэффициента**

```python
# Рассчитываем матрицу корреляции phi_k
# Указываем интервальные переменные для корректной обработки
phik_corr_matrix = df_phik.phik_matrix(interval_cols=['rating', 'seats'])

# Извлекаем корреляции только с рейтингом
rating_corr = phik_corr_matrix['rating'].sort_values(ascending=False)

print("Матрица корреляций φ-коэффициента (только с рейтингом):")
display(rating_corr)

# Визуализируем тепловую карту
plt.figure(figsize=(8, 6))
sns.heatmap(phik_corr_matrix[['rating']].sort_values('rating', ascending=False),
            annot=True, fmt='.2f', cmap='coolwarm', center=0, vmin=0, vmax=0.5,
            square=True, linewidths=0.5, linecolor='white')
plt.title('Корреляция φ (phi-k) рейтинга с другими признаками', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```

**Промежуточный вывод 6.2:**

Все корреляции с рейтингом оказались очень слабыми `(φ < 0.3). `
Это говорит о том, что ни один из рассмотренных факторов (категория, район, сетевой статус, 
количество мест, цена, круглосуточная работа) не имеет сильной прямой связи с рейтингом заведения.

**6.3. Анализ самой сильной связи (рейтинг и ценовая категория)**

```python
# Находим признак с самой сильной корреляцией (исключая сам рейтинг)
corr_without_rating = rating_corr.drop('rating')
strongest_corr = corr_without_rating.idxmax()
strongest_value = rating_corr[strongest_corr]

print(f"Самая сильная связь: `{strongest_corr}` (φ = `{strongest_value:.3f}`)")

# Визуализируем связь рейтинга с ценовой категорией
df_price_filtered = df[df['price'] != 'не указано'].copy()

plt.figure(figsize=(12, 6))
order = df_price_filtered.groupby('price', observed=True)['rating'].median().sort_values(ascending=False).index
sns.boxplot(data=df_price_filtered, x='price', y='rating', order=order, hue='price', palette='viridis')
plt.title('Зависимость рейтинга от ценовой категории', fontsize=14, fontweight='bold')
plt.xlabel('Ценовая категория', fontsize=12)
plt.ylabel('Рейтинг', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Дополнительная статистика
price_rating_stats = df_price_filtered.groupby('price', observed=True)['rating'].agg(['mean', 'median', 'count']).round(3)
print("Статистика рейтингов по ценовым категориям:")
display(price_rating_stats)
```

**8.4. Сравнение среднего чека по административным округам**

```python
# 8.4. Ранжирование округов по ценам и анализ влияния удаленности от центра

# Используем правильную переменную df_price_filtered (созданную в задаче 8.1)
# Если эта переменная не определена, создадим ее заново
if 'df_price_filtered' not in locals():
    df_price_filtered = df[df['middle_avg_bill'].notna()].copy()
    print("Переменная df_price_filtered создана заново")

# Группируем по округам и считаем статистику
avg_bill_by_district = (df_price_filtered.groupby('district', observed=True)['middle_avg_bill']
                        .agg(['mean', 'median', 'count', 'std', 'min', 'max'])
                        .sort_values('median', ascending=False)
                        .round(0)  # Округляем для читаемости
                        .astype(int) # Преобразуем в целые числа для красоты
                        )

print("Рейтинг административных округов по медианному среднему чеку (в рублях):")
display(avg_bill_by_district[['median', 'mean', 'count']])

# Визуализация с цветовой дифференциацией по принципу "ближе к центру - дальше от центра"
plt.figure(figsize=(14, 7))

# Создаем список цветов: Центр (красный), Ближний пояс (оранжевый), Дальний пояс (синий)
colors = []
for district in avg_bill_by_district.index:
    if 'Центральный' in district:
        colors.append('red')
    elif any(name in district for name in ['Западный', 'Северный', 'Южный', 'Восточный']):
        # Округа, граничащие с центром или близкие к нему
        colors.append('orange')
    else:
        # Округа дальней периферии
        colors.append('skyblue')

bars = plt.bar(avg_bill_by_district.index, avg_bill_by_district['median'], 
               color=colors, edgecolor='black', alpha=0.8)

# Добавляем значения на столбцы
for bar, val in zip(bars, avg_bill_by_district['median']):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 15, 
             f'{val}', ha='center', va='bottom', fontsize=9)

plt.xlabel('Административный округ', fontsize=12)
plt.ylabel('Медианный средний чек, руб.', fontsize=12)
plt.title('Влияние удаленности от центра на цены в заведениях Москвы', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Добавляем легенду для пояснения цветов
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='red', edgecolor='black', label='Центр (ЦАО)'),
                   Patch(facecolor='orange', edgecolor='black', label='Ближний пояс (граничащие с центром)'),
                   Patch(facecolor='skyblue', edgecolor='black', label='Периферия (удаленные округа)')]
plt.legend(handles=legend_elements, loc='upper right')

plt.tight_layout()
plt.show()

# Анализ разрыва между "центром" и "периферией"
central_median = avg_bill_by_district.loc['Центральный административный округ', 'median']
periphery_medians = avg_bill_by_district[avg_bill_by_district.index != 'Центральный административный округ']['median']
avg_periphery_median = periphery_medians.mean()

# Находим округ с самым низким чеком
lowest_median_district = avg_bill_by_district['median'].idxmin()
lowest_median_value = avg_bill_by_district.loc[lowest_median_district, 'median']

print(f"""
Ключевые выводы о влиянии локации на цены:

Центр vs Периферия: Медианный чек в Центральном округе ({central_median} руб.) в {central_median/avg_periphery_median:.1f} раза выше, чем в среднем по периферии ({avg_periphery_median:.0f} руб.).
Градиент цен: Наблюдается четкая закономерность: чем дальше от центра, тем ниже цены. 
    - Центр (ЦАО): {central_median} руб.
    - Ближний пояс: от {avg_bill_by_district.loc['Северный административный округ', 'median']} до {avg_bill_by_district.loc['Западный административный округ', 'median']} руб.
    - Периферия: самый низкий чек в {lowest_median_district} — всего {lowest_median_value} руб.
   Практический вывод: Для бизнеса с высоким средним чеком (премиум-сегмент) критически важна локация в центре или ближнем поясе. Для демократичных форматов периферия предлагает сниженную конкуренцию и доступную аренду, но и более низкий порог цен для клиентов.
""")
```

**Промежуточный вывод 8.4:**

1. **Лидеры по ценам:**

   - Это объясняется высокой концентрацией деловых центров, туристов и платежеспособной аудитории

2. **Аутсайдеры:** 
   - Самый низкий медианный чек — в **Юго-Восточном** `(450 руб.)` и **Северо-Восточном** `(500 руб.)` округах

**8.5. Сравнение Центрального округа с остальными**

```python
# Выделяем данные по центру и остальным округам
central_price = price_df[price_df['district'] == 'Центральный административный округ']['middle_avg_bill']
other_price = price_df[price_df['district'] != 'Центральный административный округ']['middle_avg_bill']

# Статистика по группам
central_stats = central_price.describe()
other_stats = other_price.describe()

# Рассчитываем разницу заранее
diff_count = central_stats['count'] - other_stats['count']
diff_mean = central_stats['mean'] - other_stats['mean']
diff_median = central_stats['50%'] - other_stats['50%']
diff_min = central_stats['min'] - other_stats['min']
diff_max = central_stats['max'] - other_stats['max']

print(f"""
Сравнительная статистика:


| {'Показатель':<20} | {'Центр':<15} | {'Остальные':<15} | {'Разница':<15} |
| {'-'*20:<20} | {'-'*15:<15} | {'-'*15:<15} | {'-'*15:<15} |
| {'Кол-во заведений':<20} | {central_stats['count']:<15.0f} | {other_stats['count']:<15.0f} | {diff_count:<+15.0f} |
| {'Средний чек':<20} | {f"{central_stats['mean']:.0f} руб.":<15} | {f"{other_stats['mean']:.0f} руб.":<15} | {f"{diff_mean:+.0f} руб.":<15} |
| {'Медианный чек':<20} | {f"{central_stats['50%']:.0f} руб.":<15} | {f"{other_stats['50%']:.0f} руб.":<15} | {f"{diff_median:+.0f} руб.":<15} |
| {'Минимальный чек':<20} | {f"{central_stats['min']:.0f} руб.":<15} | {f"{other_stats['min']:.0f} руб.":<15} | {f"{diff_min:+.0f} руб.":<15} |
| {'Максимальный чек':<20} | {f"{central_stats['max']:.0f} руб.":<15} | {f"{other_stats['max']:.0f} руб.":<15} | {f"{diff_max:+.0f} руб.":<15} |
""")

# Визуализация сравнения
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Полный масштаб (с выбросами)
bp1 = ax1.boxplot([central_price, other_price], tick_labels=['Центр', 'Остальные'], 
                  patch_artist=True, boxprops=dict(facecolor='lightcoral', alpha=0.7))
ax1.set_title('Сравнение среднего чека (полный масштаб)', fontsize=14, fontweight='bold')
ax1.set_ylabel('Средний чек, руб.', fontsize=12)
ax1.grid(axis='y', linestyle='--', alpha=0.3)

# Детальный вид (без выбросов)
bp2 = ax2.boxplot([central_price, other_price], tick_labels=['Центр', 'Остальные'], 
                  patch_artist=True, boxprops=dict(facecolor='lightblue', alpha=0.7),
                  showfliers=False)
ax2.set_title('Сравнение среднего чека (без выбросов)', fontsize=14, fontweight='bold')
ax2.set_ylabel('Средний чек, руб.', fontsize=12)
ax2.grid(axis='y', linestyle='--', alpha=0.3)

plt.suptitle('Сравнение цен в Центральном округе и остальных районах', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
